```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# 1. 10페이지까지 돌며 명언 100개 수집 (이미지의 100 rows를 맞추기 위함)
all_quotes = []
for i in range(1, 11):
    url = f"http://quotes.toscrape.com/page/{i}/"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    quotes = [q.get_text() for q in soup.find_all('span', class_='text')]
    all_quotes.extend(quotes)

# 2. CountVectorizer 설정
# 이미지에는 'about', 'above' 같은 단어가 포함되어 있으므로 stop_words는 쓰지 않습니다.
# 또한 숫자 2, 3이 보이므로 binary=False(기본값)로 둡니다.
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(all_quotes)

# 3. 데이터프레임 생성
df_vector = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# 4. 이미지처럼 출력하기
# 코랩 환경에서 df_vector를 그냥 실행하거나 display()를 쓰면 이미지와 같은 검은색 표가 나옵니다.
df_vector
```
