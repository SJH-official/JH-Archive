## **과제 1. faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라**

- [click faker](https://faker.readthedocs.io/en/master/)

```python
!pip install faker
import pandas as pd
from faker import Faker

# Faker 설정 (한국어 데이터가 필요하다면 'ko_KR' 사용)
fake = Faker('ko_KR')

# 1. 클래스 정의
class UserProfile:
    def __init__(self, name, age, address, phone, job):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone
        self.job = job

# 2. 1000명의 데이터 생성 (클래스 인스턴스 리스트)
user_list = []
for _ in range(1000):
    user = UserProfile(
        name=fake.name(),
        age=fake.random_int(min=20, max=60),
        address=fake.address(),
        phone=fake.phone_number(),
        job=fake.job()
    )
    user_list.append(user)

# 3. 클래스 인스턴스 리스트를 pandas 데이터프레임으로 변환
df = pd.DataFrame([vars(u) for u in user_list])

# 결과 출력
display(df.head())
print(f'총 생성된 데이터 수: {len(df)}개')
```

## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class User:
    def __init__(self, username):
        # 1. 속성(Attributes) 초기화
        self.username = username
        self.following = 0  # 내가 쫓아가는 대상 수
        self.followers = 0  # 나를 쫓아오는 대상 수
        print(f"[{self.username}] 캐릭터가 생성되었습니다.")

    def following_mode(self, target):
        # 2. 속성 변경 (쫓아가는 행위)
        # 내가 누군가를 쫓아가면(following), 상대방은 추격자(follower)가 생기는 것
        self.following += 1
        target.followers += 1
        print(f"🏃 {self.username}이(가) {target.username}을(를) 추격합니다!")
        print(f"   (결과 -> {self.username}의 Following: {self.following} | {target.username}의 Followers: {self.followers})")

# --- 실행부 ---

# Tom과 Jerry 객체 생성
tom = User("Tom")
jerry = User("Jerry")

print("\n--- 추격전 시작! ---")

# 10번 반복해서 Tom이 Jerry를 쫓아가게 만들기
for i in range(1, 11):
    print(f"[{i}회차]")
    tom.following_mode(jerry)

print("\n--- 최종 결과 ---")
print(f"Tom의 최종 팔로잉 수: {tom.following}")
print(f"Jerry의 최종 팔로워 수: {jerry.followers}")
```
